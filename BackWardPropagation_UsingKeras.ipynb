```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[5,12,7]], columns= ['cgpa', 'profile_score','lpa'])
```

```python
df
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(2, activation='linear', input_dim=2))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python

new_weights = [np.array([[0.1, 0.1],
[0.1, 0.1 ]], dtype = np.float32),
np.array([0., 0.], dtype=np.float32), np.array([[0.1],
[ 0.1]], dtype=np.float32), np.array([0.], dtype=np.float32)]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="mean_squared_error", optimizer=optimizer)
```

```python
model.fit(df.iloc[:,[0,1]].values, df['lpa'].values, epochs=190,verbose = 1, batch_size=1)
```
