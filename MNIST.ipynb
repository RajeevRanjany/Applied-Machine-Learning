28 * 28 : 784 input and 10 outputs and whose probability will be maximum that will be answer. 1 Input for each pixels so weight and biases will be huge

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train,y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train / 255
X_test = X_test /255
```

```python
X_train[0]
```

Now we have to build ANN Model

```python

model = Sequential()
# first flatten the input from 2d to 1d
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu')) # to optimize this
model.add(Dense(10, activation='softmax'))
# model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# in this loss function we don't need to do one hot encoding
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=25,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
